```python
import numpy as np
import matplotlib.pyplot as plt
from pymoo.util.nds.non_dominated_sorting import NonDominatedSorting

# Extract objective values from the optimization result
F = res.F  # Shape (n_solutions, 2), columns: [1-KGE_Yield, 1-KGE_Irrigation]

# Get Pareto front solutions (Rank 0 from non-dominated sorting)
nds = NonDominatedSorting().do(F)
pareto_front = F[nds[0]]

# Scatter plot of all solutions
plt.figure(figsize=(8, 6))
plt.scatter(F[:, 0], F[:, 1], color='gray', alpha=0.5, label="All Solutions")

# Highlight the Pareto front solutions
plt.scatter(pareto_front[:, 0], pareto_front[:, 1], color='red', label="Pareto Front")

# Labels and title
plt.xlabel("1 - KGE_Yield (Lower is Better)")
plt.ylabel("1 - KGE_Irrigation (Lower is Better)")
plt.title("Pareto Front for AquaCrop Calibration")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
