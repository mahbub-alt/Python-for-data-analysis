```python
# Step 1: Calculate residuals
df["Irrigation_Residual"] = (df["Reported_Irrigation"] - df["Simulated_Irrigation"]).abs()
df["Yield_Residual"] = (df["Reported_Yield"] - df["Simulated_Yield"]).abs()
df["ET_Residual"]    =  (df['Ensemble_ET'] - df['ET_aqua']).abs()


# Step 2: Normalize residuals using Min-Max scaling
def min_max_scaling(series):
    return (series - series.min()) / (series.max() - series.min())

df["Normalized_Irrigation_Residual"] = min_max_scaling(df["Irrigation_Residual"])
df["Normalized_Yield_Residual"] = min_max_scaling(df["Yield_Residual"])
df["Normalized_ET_Residual"] = min_max_scaling(df["ET_Residual"])
# Step 3: Calculate inverse variance weights (if we use just raw residual then variance seems
#too high for irrigation compared to yield, probably because of their different scale and high diference 
#inthe case of irrigation)

# var_irrigation = np.var(df["Irrigation_Residual"], ddof=1)
# var_yield = np.var(df["Yield_Residual"], ddof=1)

# using normalized var ( works the best)
var_irrigation = np.var(df["Normalized_Irrigation_Residual"] , ddof=1)
var_yield = np.var(df["Normalized_Yield_Residual"], ddof=1)


weight_irrigation = 1 / var_irrigation
weight_yield = 1 / var_yield

# Step 4: Compute the weighted mean of normalized residuals
mean_normalized_irrigation_residual = df["Normalized_Irrigation_Residual"].mean()
mean_normalized_yield_residual = df["Normalized_Yield_Residual"].mean()

# Objective function
objective_function = (
    weight_irrigation * mean_normalized_irrigation_residual +
    weight_yield * mean_normalized_yield_residual
)

# Output results
print("Weight (Irrigation):", weight_irrigation)
print("Weight (Yield):", weight_yield)
print(" Irrigation contribution:",weight_irrigation * mean_normalized_irrigation_residual)
print("Yield contribution:",  weight_yield *mean_normalized_yield_residual)
print("Objective Function Value:", objective_function)
print("Manual weight (60/40):",4 * mean_normalized_irrigation_residual+6 *mean_normalized_yield_residual )
```

```python
def obj_test (df, no_ET =True):
    df["Irrigation_Residual"] = (df["Reported_Irrigation"] - df["Simulated_Irrigation"]).abs()
    df["Yield_Residual"] = (df["Reported_Yield"] - df["Simulated_Yield"]).abs()



    # Step 2: Normalize residuals using Min-Max scaling
    def min_max_scaling(series):
        return (series - series.min()) / (series.max() - series.min())

    df["Normalized_Irrigation_Residual"] = min_max_scaling(df["Irrigation_Residual"])
    df["Normalized_Yield_Residual"] = min_max_scaling(df["Yield_Residual"])


    # Step 3: Compute mean of normalized residuals
    mean_normalized_irrigation_residual = df["Normalized_Irrigation_Residual"].mean()
    mean_normalized_yield_residual = df["Normalized_Yield_Residual"].mean()

    if no_ET:
        weight_yield = 6
        weight_irrigation =4


        # Objective function
        fitness = (
            weight_irrigation * mean_normalized_irrigation_residual +
            weight_yield * mean_normalized_yield_residual
        )

        # Output results
        print("Mean Normalized Irrigation Residual:", mean_normalized_irrigation_residual)
        print("Mean Normalized Yield Residual:", mean_normalized_yield_residual)
        print(" Irrigation contribution:",weight_irrigation * mean_normalized_irrigation_residual)
        print("Yield contribution:",  weight_yield *mean_normalized_yield_residual)
        print("Objective Function Value:", fitness)

    else:
        df["ET_Residual"] =  (df['Ensemble_ET'] - df['ET_aqua']).abs()
        df["Normalized_ET_Residual"] = min_max_scaling(df["ET_Residual"])
        mean_normalized_ET_residual = df["Normalized_ET_Residual"].mean()
        
        weight_yield = 5
        weight_irrigation =3
        weight_et  = 2

        # Objective function
        fitness = (
            weight_irrigation * mean_normalized_irrigation_residual +
            weight_yield * mean_normalized_yield_residual +
            weight_et * mean_normalized_ET_residual
        )

        # Output results
        print("Mean Normalized Irrigation Residual:", mean_normalized_irrigation_residual)
        print("Mean Normalized Yield Residual:", mean_normalized_yield_residual)
        print("ET_resid:", mean_normalized_ET_residual)
        print(" Irrigation contribution:",weight_irrigation * mean_normalized_irrigation_residual)
        print("Yield contribution:",  weight_yield *mean_normalized_yield_residual)
        print("ET contribution:",  weight_et * mean_normalized_ET_residual)
        print("Objective Function Value:", fitness)
```

```python
def obj_func (param, no_ET =True, Train = True):
    
    """
    you can remove any parameter from "for_objf" function
    to avaoid calibration of that parameter. for that lb and up adjustment is necessary.
    """

    smt = [param[0], param[1], param[2], param[3]]
    CCx = param[4]
    WP = param[5] 
    Kcb = param[6]
    HI0 = param[7]
    a_HI = param[8]
    Zmax = param[7]

# , a_HI = a_HI, Zmax = Zmax
    print("smt:", smt)
    print("CCx:", CCx)
    print("WP:",  WP)
    print("Kcb:", Kcb)
    print("HI0:", HI0)
    print("a_HI:", a_HI)
    print("Zmax:",Zmax)
    
    all_ET_df, all_yld_df, simul_reported = for_objf(smt =smt, CCx =CCx, WP = WP, Kcb = Kcb,
                                                     HI0 = HI0, a_HI = a_HI, Zmax = Zmax,
                                                     no_ET = no_ET,Train =Train)
    
    

    simul_reported.rename(columns={"Seasonal irrigation (mm)": "Simulated Irrigation", "Yield (tonne/ha)": "Simulated Yield"}, inplace =True)
    
    Reported_Yield = simul_reported['Reported_Yield']
    Simulated_Yield = simul_reported['Simulated Yield']
    Reported_Irrigation = simul_reported['Reported_Irrigation']
    Simulated_Irrigation = simul_reported['Simulated Irrigation']
    
    yld_res = ((Reported_Yield.values - Simulated_Yield.values) ** 2) 
    irri_res = ((Reported_Irrigation.values - Simulated_Irrigation.values) ** 2) 
    
    yield_var = np.nanvar(Reported_Yield.values - Simulated_Yield.values)
    irrig_var = np.nanvar(Reported_Irrigation - Simulated_Irrigation)
    
    if  no_ET:

        fitness = np.nanmean((yld_res / yield_var)) + np.nanmean ((irri_res / irrig_var))
    if no_ET == False:
        Reported_ET = simul_reported ['ET_aqua']
        Simulated_ET = simul_reported ['Ensemble_ET']
        ET_res = ((Reported_ET.values - Simulated_ET.values) ** 2)
        ET_var = (np.nanvar(Reported_ET.values - Simulated_ET.values))
        print("yld_fit:",np.nanmean((((Reported_Yield.values - Simulated_Yield.values) ** 2) * (1 / yield_var))))
        print ("irri_fit:",np.nanmean((((Reported_Irrigation.values - Simulated_Irrigation.values) ** 2) * (1 / irrig_var))))
        print("ET_fit:",np.nanmean((((Reported_ET.values - Simulated_ET.values) ** 2) * (1/ET_var)))*.55)
        fitness = np.nanmean ((yld_res / yield_var)) + np.nanmean((irri_res / irrig_var)) + np.nanmean((ET_res/ET_var))

        
    print('len:',len(simul_reported))
    ###### obj function (mae)

    print("\033[91m" + "loss:", fitness, "\033[0m")
    print("------------------------")
    
    return fitness
```
